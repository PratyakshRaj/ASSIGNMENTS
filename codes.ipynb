```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import stats as stats
import pandas as pd
```

# Q1-

```python
mean=1.5
std=0.5
```

# Random samples

```python
x = stats.norm(mean,std).rvs(1000)
```

# PDF Graph

```python
l_s = np.linspace(mean - 3*std, mean + 3*std, 1000)
plt.plot(l_s, stats.norm.pdf(l_s, mean, std))
plt.show()
```

```python
q25,q50,q75 = np.percentile(x,[25,50,75])
```

# MAD

```python
mad=stats.median_abs_deviation(x)
mad
```

```python
sigma_G = 0.7413*(q75-q25)
sigma_G
```

# Standard Diviation

```python
std_new=mad*1.482
std_new
```

# Mean

```python
mean= np.mean(x)
mean
```

# Variance

```python
var=np.var(x)
var
```

# Skewness

```python
skew=stats.skew(x)
skew
```

# Kurtosis

```python
kur=stats.kurtosis(x)
kur
```

# Q2-

```python
l_s = np.linspace(-4, 4, 1000)
plt.plot(l_s, stats.cauchy.pdf(l_s, 0, 1.5),ls='dashed',label='cauchy')
plt.plot(l_s, stats.norm.pdf(l_s, 0, 1.5), label='Guassian')
plt.legend()
plt.show()
```

# Q3-

```python
l_s = np.linspace(5 - 3*(5**0.5), 5 + 3*(5**0.5), 1000)
plt.plot(l_s, stats.poisson.pmf(l_s,5),ls='dashed',label='poisson')
plt.plot(l_s, stats.norm.pdf(l_s, 5,5**0.5), label='Guassian')
plt.legend()
plt.show()

```
